# Clustering The Tweets Using Affinitiy Propagation

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import text 
from sklearn.metrics.pairwise import pairwise_distances
import numpy
import networkx as nx

def getClusters(day, time_window):
    tweets = get_tweets(day, time_window)
    stop_words = text.ENGLISH_STOP_WORDS
    #we need to remove the url to compare the similarity
    corps = []
    for t in tweets:
        content = t["text"]
        #print t.keys()
        try:
            urls = [u["url"] for u in t['entities']["urls"]]
            for url in urls:
                content = content.replace(url,"")
        except:
            pass
        content = content.replace("\n\n","").replace("\r","")
        corps.append(content)
        
    
    vectorizer = CountVectorizer(stop_words=stop_words)
    X = vectorizer.fit_transform(corps)
    similarity = 1 - pairwise_distances(X, metric='cosine')
    
    #construct a tweets network to get clusters
    tweet_graph = nx.Graph()
    us, vs = np.nonzero(similarity>0.6)
    for i in range(len(us)):
        start = us[i]
        end = vs[i]
        if start == end:
            continue
        tweet_graph.add_node(us[i], value=corps[us[i]])
        tweet_graph.add_node(vs[i], value=corps[vs[i]])
        tweet_graph.add_edge(us[i], vs[i])

    components = nx.connected_components(tweet_graph)
    for i, nodes in enumerate(components[:3]):
        print 'Cluster[%d]--Size: %d' % (i, len(nodes))
        #using the nodes with hightest sum similarity as examplar
        cluster_sim = 0
        cluster_idx = -1
        for node in nodes:
            tmp_sum = 0
            for node2 in nodes:
                if node2 != node:
                    tmp_sum += similarity[node][node2]
            if tmp_sum > cluster_sim:
                cluster_sim = tmp_sum
                cluster_idx = node
                    
        print "\tExemplar:  %s" % tweet_graph.node[cluster_idx]["value"]
```
