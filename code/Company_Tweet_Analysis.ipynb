## Description

We downloaded the tweets for Monolithic Power (MPWR), Boyd Gaming (BYD), Sapient (SAPE) since 2007. The tweets amount related to MPWR is relative small(only 2000). 
The tweets for SAPE are relative noised, a lot of them just contain the word sapient and have nothing to do with the company Sapient(SAPE). 
The quality of tweets for Boyd Gaming (BYD) is relative good both in precision and volume(around 20000). So I focus on the BYD data set. Since Since Twitter was founded in mid 2006, there are few tweets in 2006 and 2007 and so I only use the data from 2010 to 2013. 
I collected a list of important dates for each company. The important dates are the dates when the company release 10-K, 10-Q and 8-K reports.
The 10-K is the annual financial report, 10-Q is the quarterly financial report and 8-K is the report used to announce important events to the company. 
We ploted events along with company stock price and tweet volume. The red and blue points in the charts representing the events.
There are two interesting findings:
1> The peak of tweets volume can capture most of the important events. 
2> Most of the time, the 10-Q events are preceded by the 8-K events

We also analyzed count the retweet information. But the amount of retweet are relative small, most of the tweets are original. Only in 2013, the retweets information can capture the important events. You can find the figure named BYD_Retweets in below cells.

To analyze the relation beween price fluctuation between the important event days, we ploted zscore30 of the price along with the tweets volume. We can find that there exists some signicant price changes around the event days. The zscore figures are named by BYD_Tweets_zscore30.

# Code for the analysis

```python


import json
from glob import glob
import json
import datetime
import time
from dateutil import parser
import urllib2
import urllib
from bs4 import BeautifulSoup

"""
work steps:
1> getting the stock price of these companies from yahoo
2> get the annual report date of these companies from EDGAR
3> compute the daily tweet count for these companies
4> find the entities in the tweets
5> find who are the guies most tweet about these companies
6> which tweets of these companies has been retweeted many times(the retweets also capture some events, but it seems that 
people people only talk it and in the same day)
7> compute the zscore of the stock market to check the 
"""

mpwr_files = glob("/home/weiwang/workspace/data/topsy_50/MPWR*")
byd_files = glob("/home/weiwang/workspace/data/topsy_50/BYD*")
sape_files = glob("/home/weiwang/workspace/data/topsy_50/SAPE*")
date_format = "%Y-%m-%d"

#count the total tweets volumn by day
def group_by_day(files):
    result = {}
    for f in files:
        try:
            with open(f) as tf:
                for l in tf:
                    t = json.loads(l)
                    dt = parser.parse(t['created_at']).strftime(date_format)
                    if dt in result:
                        result[dt] = result[dt] + 1
                    else:
                        result[dt] = 1
        except:
            print f
    days = sorted(result.keys())
    values = [result[d] for d in days]
    return days, values, result

#count the retweets volumn by day
def group_retweet_by_day(files):
    result = {}
    for f in files:
        try:
            with open(f) as tf:
                for l in tf:
                    t = json.loads(l)
                    dt = parser.parse(t['created_at']).strftime(date_format)
                    if "retweeted_status" in t and t["retweeted_status"] != {}:
                        if dt in result:
                            result[dt] = result[dt] + 1
                        else:
                            result[dt] = 1
        except:
            print f
    days = sorted(result.keys())
    values = [result[d] for d in days]
    return days, values, result
    
#get the quotes information from Yahoo finance
def get_quotes(symbol, start, end):
    from matplotlib.finance import quotes_historical_yahoo
    s_year, s_month, s_day = map(int, start.split('-'))
    e_year, e_month, e_day = map(int, end.split('-'))
    date1 = datetime.date( s_year, s_month, s_day)
    date2 = datetime.date( e_year, e_month, e_day)
    quotes = quotes_historical_yahoo(symbol, date1, date2, adjusted=True)
    if len(quotes) == 0:
        print "No quotes data"
        return
    days = [matplotlib.dates.num2date(q[0]).strftime("%Y-%m-%d") for q in quotes]
    closes = [q[2] for q in quotes]
    return {days[i]: closes[i] for i in range(len(days))}

#helper function to compuate the zscore 
def compute_zscore(datas, window):
    zscores = [0]
    for i in range(1, len(datas)):
        cur = datas[i]
        previous = datas[0 if i - window <=0 else i - window:i]
        try:
            pre_std = np.std(previous)
            if np.abs(pre_std) < 0.001:
                zscore = 0.0
            else:
                zscore = (cur - np.mean(previous))/np.std(previous)
        except:
            zscore = 0.0
        zscores.append(zscore)
    return zscores

#compute the stock price zscore value
def price_zscore(quotes, window):
    days = sorted(quotes.keys(), reverse=False)
    prices = [quotes[d] for d in days]
    zscores = compute_zscore(prices, window)
    return dict(zip(days, zscores))

"""
Extract import days from EDGAR:
consists of 10-Q(quarterly report), 10-K(Annual report), 8-K(important event)
"""

byd_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0000906553&type=8-K%25&dateb=&owner=exclude&count=40"
BASIC_URL = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=%s&type=%s&dateb=&owner=exclude&count=40"

#helper function to ingest event days from the html soup file
def get_days(soup):
    obs = soup.find_all(class_="tableFile2")[0].find_all('tr')
    days = [o.find_all('td')[3].text for o in obs[1:]]
    return sorted(days, reverse=True)

#helper function ingest event days from url
def ingest_event_days(url):
    event_days = []
    soup = BeautifulSoup(urllib2.urlopen(url))
    next_flag = soup.find_all("input", attrs={"value": "Next 40"})
    start = 0
    event_days = event_days + get_days(soup)
    while len(next_flag) > 0:
        start = start + 40
        next_url = "%s&start=%d" % (url, start)
        soup = BeautifulSoup(urllib2.urlopen(next_url))
        event_days = event_days + get_days(soup)
        next_flag = soup.find_all("input", attrs={"value": "Next 40"})
    return event_days

#ingest the event day by cik of the company
def ingest(cik):
    day_events = {}
    event_types = ["8-K", "10-Q", "10-K"]
    for event_type in event_types:
        url = BASIC_URL % (cik, event_type)
        days = ingest_event_days(url)
        day_events.update({d:event_type for d in days})
    return day_events

#ingest the quarterly report day
def ingest_quarterly_days(url):
    soup = BeautifulSoup(urllib2.urlopen(url))
    obs = soup.find_all(class_="tableFile2")[0].find_all('tr')
    days = [o.find_all('td')[3].text for o in obs[1:]]
    return sorted(days)

"""
Plot the tweets, price along with all important event days(10-K. 10-Q, 8-K) by year
"""
def plot_by_event_year(symbol, days, datas, events, quotes, year, plot_type):
    start_index = -1
    end_index = -1
    start_found = False
    end_found = False
    for i, d in enumerate(days):
        if d[0:4] == str(year) and not start_found:
            start_index = i
            start_found = True
        if d[0:4] == str(year + 1):
            end_index = i - 1
            end_found = True
            break
    if not end_found:
        end_index = i
    if not start_found:
        print "no data"
        return
    
    tweets = dict(zip(days,datas))
    fig = figure(figsize(15, 6), dpi=80)
    ax = gca()
    ax2 = ax.twinx()
    #ax.set_yscale('log')
    num_days = [matplotlib.dates.date2num(parser.parse(d)) for d in days[start_index:end_index + 1]]
    
    #filter the scatter days by year
    s_day = "%d-01-01" % year
    e_day = "%d-12-31" % year
    event_days = [d for d in events.keys() if d >= s_day and d <= e_day]
    event_count_values = [tweets.get(d,0) for d in event_days]
    num_event_days = [matplotlib.dates.date2num(parser.parse(d)) for d in event_days]
    
    #filter the historical price by year
    quote_days = sorted([d for d in quotes.keys() if d >= s_day and d <= e_day])
    quote_values = [quotes[d] for d in quote_days]
    
    num_quote_days = [matplotlib.dates.date2num(parser.parse(d)) for d in quote_days]
    event_quote_values = [quotes.get(d) for d in event_days if d in quotes]
    num_quote_event_days =[matplotlib.dates.date2num(parser.parse(d)) for d in event_days if d in quotes]
    
    #plot the data
    l1 = ax.plot_date(num_days, datas[start_index:end_index + 1], '-', color='blue', linewidth=1.5, label="Tweets")        
    l2 = ax2.plot_date(num_quote_days, quote_values, '-', color='red', linewidth=1.5, label="price")
    
    ax.set_ylim(ymin=0)
#     if plot_type == "quote":
#         ax2.set_ylim(0,19)
    
    ax.scatter(num_event_days, event_count_values, 50, color='blue')
    ax2.scatter(num_quote_event_days, event_quote_values, 50, color='red')
    
    #ax.annotate()
    for i in range(len(event_days)):
        ax.annotate(events[event_days[i]], 
                    xy=(num_event_days[i], event_count_values[i]), 
                    xycoords='data',
                    textcoords='offset points',
                    xytext=(0.05, 50),
                    arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))
    
    months = MonthLocator()
    days = DayLocator()
    monthsFmt = DateFormatter("%Y-%m-%d")
    ax.xaxis.set_major_locator(months)
    ax.xaxis.set_major_formatter(monthsFmt)
    ax.xaxis.set_minor_locator(days)
    ax.grid(True)
    ax.xaxis.set_ticks_position('bottom')
    ax.autoscale_view()
    fig.autofmt_xdate()
    lns = l1 + l2
    lables = [l.get_label() for l in lns]
    ax.legend(lns, lables, loc="upper left")
    ax.set_title(symbol)

def plot_price(quotes, name):
    days = sorted(quotes.keys())
    num_quote_days = [matplotlib.dates.date2num(parser.parse(d)) for d in days]
    event_quote_values = [quotes.get(d) for d in days]
    fig = figure(figsize(15, 6), dpi=80)
    ax = gca()
    ax.plot_date(num_quote_days, event_quote_values, '-', color='black', linewidth=1.5, label="Stock Quote")
    months = MonthLocator()
    days = DayLocator()
    monthsFmt = DateFormatter("%Y-%m-%d")
    ax.xaxis.set_major_locator(months)
    ax.xaxis.set_major_formatter(monthsFmt)
    ax.xaxis.set_minor_locator(days)
    ax.grid(True)
    ax.xaxis.set_ticks_position('bottom')
    ax.autoscale_view()
    fig.autofmt_xdate()
    ax.set_title("Stock Quote for %s" % name)

def process(company):
    """Merger all the steps together to analyze"""
    tweets_files = glob("/home/weiwang/workspace/data/topsy_50/%s" % (company.upper() + "*"))
    #quarterly report url for the three companies
    mpwr_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0001280452&type=10-Q&dateb=&owner=exclude&count=80"
    byd_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0000906553&type=10-Q&dateb=&owner=exclude&count=40"
    sape_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0001008817&type=10-Q&dateb=&owner=exclude&count=40"
    
    url_map = {"mpwr": mpwr_url, "byd":byd_url, "sape":sape_url }
    
    cik_map = {"byd": "0000906553", "sape":"0001008817", "mpwr":"0001280452"}
    report_day_url = url_map[company]
    
    tweet_days, tweet_counts, t_com_dic = group_by_day(tweets_files) #group the overall tweets count by day
    retweet_days, retweet_counts, rt_com_dic = group_retweet_by_day(tweets_files) #group the overall retweets count by day
    quotes = get_quotes(company.upper(), "2009-01-01", "2013-12-31") #ingest the quotes from yahoo finance
    quote_z30 = price_zscore(quotes, 30)  #compute the z30
    report_days = ingest_quarterly_days(report_day_url) #ingest report days
    events = ingest(cik_map[company]) #ingest the events days
    
    #plot the price data
    plot_price(quotes, company.upper())
    
    years = [2013, 2012, 2011, 2010]  #the year to print
    #plot the events along with the tweets volume and price
    for year in years:
        plot_by_event_year("%s_Tweets vs Normalized Quote" % company.upper(), tweet_days, tweet_counts, events, quotes, year, "quote")
            
    #plot the events along with the tweet volume and zscore30
    for year in years:
        plot_by_event_year("%s_Tweets vs Quote_Zscore30" % company.upper(), tweet_days, tweet_counts, events, quote_z30, year, "zscore")
        
    #plot the events along with the retweet volume and price
    for year in years:
        plot_by_event_year("%s_Retweets vs Normalized Quote" % company.upper(), retweet_days, retweet_counts, events, quotes, year, "quote")

 
```
