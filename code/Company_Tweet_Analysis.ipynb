```python


import json
from glob import glob
import json
import datetime
import time
from dateutil import parser
import urllib2
import urllib
from bs4 import BeautifulSoup
import unicodedata
import pandas as pd
from IPython.display import HTML
import re
from nltk.corpus import stopwords
from collections import Counter
import nltk

"""
work steps:
1> getting the stock price of these companies from yahoo
2> get the annual report date of these companies from EDGAR
3> compute the daily tweet count for these companies
4> find the entities in the tweets
5> find who are the guies most tweet about these companies
6> which tweets of these companies has been retweeted many times(the retweets also capture some events, but it seems that 
people people only talk it and in the same day)
7> compute the zscore of the stock market to check the 
"""

mpwr_files = glob("/home/weiwang/workspace/data/topsy_50/MPWR*")
byd_files = glob("/home/weiwang/workspace/data/topsy_50/BYD*")
sape_files = glob("/home/weiwang/workspace/data/topsy_50/SAPE*")
date_format = "%Y-%m-%d"
BASIC_URL = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=%s&type=%s&dateb=&owner=exclude&count=40"

#count the total tweets volumn by day
def group_by_day(files):
    result = {}
    for f in files:
        try:
            with open(f) as tf:
                for l in tf:
                    t = json.loads(l)
                    dt = parser.parse(t['created_at']).strftime(date_format)
                    if dt in result:
                        result[dt] = result[dt] + 1
                    else:
                        result[dt] = 1
        except:
            print f
    days = sorted(result.keys())
    values = [result[d] for d in days]
    return days, values, result

#count the retweets volumn by day
def group_retweet_by_day(files):
    result = {}
    for f in files:
        try:
            with open(f) as tf:
                for l in tf:
                    t = json.loads(l)
                    dt = parser.parse(t['created_at']).strftime(date_format)
                    if "retweeted_status" in t and t["retweeted_status"] != {}:
                        if dt in result:
                            result[dt] = result[dt] + 1
                        else:
                            result[dt] = 1
        except:
            print f
    days = sorted(result.keys())
    values = [result[d] for d in days]
    return days, values, result
    
#get the quotes information from Yahoo finance
def get_quotes(symbol, start, end):
    from matplotlib.finance import quotes_historical_yahoo
    s_year, s_month, s_day = map(int, start.split('-'))
    e_year, e_month, e_day = map(int, end.split('-'))
    date1 = datetime.date( s_year, s_month, s_day)
    date2 = datetime.date( e_year, e_month, e_day)
    quotes = quotes_historical_yahoo(symbol, date1, date2, adjusted=True)
    if len(quotes) == 0:
        print "No quotes data"
        return
    days = [matplotlib.dates.num2date(q[0]).strftime("%Y-%m-%d") for q in quotes]
    closes = [q[2] for q in quotes]
    return {days[i]: closes[i] for i in range(len(days))}

#helper function to compuate the zscore 
def compute_zscore(datas, window):
    zscores = [0]
    for i in range(1, len(datas)):
        cur = datas[i]
        previous = datas[0 if i - window <=0 else i - window:i]
        try:
            pre_std = np.std(previous)
            if np.abs(pre_std) < 0.001:
                zscore = 0.0
            else:
                zscore = (cur - np.mean(previous))/np.std(previous)
        except:
            zscore = 0.0
        zscores.append(zscore)
    return zscores

#compute the stock price zscore value
def price_zscore(quotes, window):
    days = sorted(quotes.keys(), reverse=False)
    prices = [quotes[d] for d in days]
    zscores = compute_zscore(prices, window)
    return dict(zip(days, zscores))

"""
Extract import days from EDGAR:
consists of 10-Q(quarterly report), 10-K(Annual report), 8-K(important event)
"""
#helper function to ingest event days from the html soup file
def get_days(soup):
    obs = soup.find_all(class_="tableFile2")[0].find_all('tr')
    days = [o.find_all('td')[3].text for o in obs[1:]]
    return sorted(days, reverse=True)

#helper function ingest event days from url
def ingest_event_days(url):
    event_days = []
    soup = BeautifulSoup(urllib2.urlopen(url))
    next_flag = soup.find_all("input", attrs={"value": "Next 40"})
    start = 0
    event_days = event_days + get_days(soup)
    while len(next_flag) > 0:
        start = start + 40
        next_url = "%s&start=%d" % (url, start)
        soup = BeautifulSoup(urllib2.urlopen(next_url))
        event_days = event_days + get_days(soup)
        next_flag = soup.find_all("input", attrs={"value": "Next 40"})
    return event_days

#ingest the event day by cik of the company
def ingest(cik):
    day_events = {}
    event_types = ["8-K", "10-Q", "10-K"]
    for event_type in event_types:
        url = BASIC_URL % (cik, event_type)
        days = ingest_event_days(url)
        day_events.update({d:event_type for d in days})
    return day_events

#ingest the quarterly report day
def ingest_quarterly_days(url):
    soup = BeautifulSoup(urllib2.urlopen(url))
    obs = soup.find_all(class_="tableFile2")[0].find_all('tr')
    days = [o.find_all('td')[3].text for o in obs[1:]]
    return sorted(days)

#get the customized event day
def get_custormized_days(company):
    eventDays = json.load(open('../data/eventDays.json'))
    company_days = eventDays.get(company, None)
    result = {day:day[5:] for day in company_days}
    return result
    
"""
Plot the tweets, price along with all important event days(10-K. 10-Q, 8-K) by year
"""
def plot_by_event_year(symbol, days, datas, events, quotes, year, plot_type):
    start_index = -1
    end_index = -1
    start_found = False
    end_found = False
    for i, d in enumerate(days):
        if d[0:4] == str(year) and not start_found:
            start_index = i
            start_found = True
        if d[0:4] == str(year + 1):
            end_index = i - 1
            end_found = True
            break
    if not end_found:
        end_index = i
    if not start_found:
        print "no data"
        return
    
    tweets = dict(zip(days,datas))
    fig = figure(figsize(15, 6), dpi=80)
    ax = gca()
    ax2 = ax.twinx()
    #ax.set_yscale('log')
    num_days = [matplotlib.dates.date2num(parser.parse(d)) for d in days[start_index:end_index + 1]]
    
    #filter the scatter days by year
    s_day = "%d-01-01" % year
    e_day = "%d-12-31" % year
    event_days = [d for d in events.keys() if d >= s_day and d <= e_day]
    event_count_values = [tweets.get(d,0) for d in event_days]
    #print "Average count: ",symbol,year, sum(event_count_values) / len(event_count_values)
    
    num_event_days = [matplotlib.dates.date2num(parser.parse(d)) for d in event_days]
    
    #filter the historical price by year
    quote_days = sorted([d for d in quotes.keys() if d >= s_day and d <= e_day])
    quote_values = [quotes[d] for d in quote_days]
    
    num_quote_days = [matplotlib.dates.date2num(parser.parse(d)) for d in quote_days]
    event_quote_values = [quotes.get(d) for d in event_days if d in quotes]
    num_quote_event_days =[matplotlib.dates.date2num(parser.parse(d)) for d in event_days if d in quotes]
    
    #plot the data
    l1 = ax.plot_date(num_days, datas[start_index:end_index + 1], '-', color='blue', linewidth=1.5, label="Tweets") 
    #print symbol,year, "Max count: ",max(datas[start_index:end_index + 1]), 'Min count:', min(datas[start_index:end_index + 1])
    l2 = ax2.plot_date(num_quote_days, quote_values, '-', color='red', linewidth=1.5, label="price")
    
    ax.set_ylim(ymin=0)
#     if plot_type == "quote":
#         ax2.set_ylim(0,19)
    
    ax.scatter(num_event_days, event_count_values, 50, color='blue')
    ax2.scatter(num_quote_event_days, event_quote_values, 50, color='red')
    
    #ax.annotate()
    for i in range(len(event_days)):
        ax.annotate(events[event_days[i]], 
                    xy=(num_event_days[i], event_count_values[i]), 
                    xycoords='data',
                    textcoords='offset points',
                    xytext=(0.05, 50),
                    arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))
    
    months = MonthLocator()
    days = DayLocator()
    monthsFmt = DateFormatter("%Y-%m-%d")
    ax.xaxis.set_major_locator(months)
    ax.xaxis.set_major_formatter(monthsFmt)
    ax.xaxis.set_minor_locator(days)
    ax.grid(True)
    ax.xaxis.set_ticks_position('bottom')
    ax.autoscale_view()
    fig.autofmt_xdate()
    lns = l1 + l2
    lables = [l.get_label() for l in lns]
    ax.legend(lns, lables, loc="upper left")
    ax.set_title(symbol)

def plot_price(quotes, name):
    days = sorted(quotes.keys())
    num_quote_days = [matplotlib.dates.date2num(parser.parse(d)) for d in days]
    event_quote_values = [quotes.get(d) for d in days]
    fig = figure(figsize(15, 6), dpi=80)
    ax = gca()
    ax.plot_date(num_quote_days, event_quote_values, '-', color='black', linewidth=1.5, label="Stock Quote")
    months = MonthLocator()
    days = DayLocator()
    monthsFmt = DateFormatter("%Y-%m-%d")
    ax.xaxis.set_major_locator(months)
    ax.xaxis.set_major_formatter(monthsFmt)
    ax.xaxis.set_minor_locator(days)
    ax.grid(True)
    ax.xaxis.set_ticks_position('bottom')
    ax.autoscale_view()
    fig.autofmt_xdate()
    ax.set_title("Stock Quote for %s" % name)

def process(company):
    """Merger all the steps together to analyze"""
    tweets_files = glob("/home/weiwang/workspace/data/topsy_50/%s" % (company.upper() + "*"))
    #quarterly report url for the three companies
    mpwr_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0001280452&type=10-Q&dateb=&owner=exclude&count=80"
    byd_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0000906553&type=10-Q&dateb=&owner=exclude&count=40"
    sape_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0001008817&type=10-Q&dateb=&owner=exclude&count=40"
    
    url_map = {"mpwr": mpwr_url, "byd":byd_url, "sape":sape_url }
    
    cik_map = {"byd": "0000906553", "sape":"0001008817", "mpwr":"0001280452"}
    report_day_url = url_map[company]
    
    tweet_days, tweet_counts, t_com_dic = group_by_day(tweets_files) #group the overall tweets count by day
    retweet_days, retweet_counts, rt_com_dic = group_retweet_by_day(tweets_files) #group the overall retweets count by day
    quotes = get_quotes(company.upper(), "2009-01-01", "2014-12-31") #ingest the quotes from yahoo finance
    quote_z30 = price_zscore(quotes, 30)  #compute the z30
    #report_days = ingest_quarterly_days(report_day_url) #ingest report days
    #events = ingest(cik_map[company]) #ingest the events days
    events = get_custormized_days(company) #ingest the customized days
    #plot the price data
    plot_price(quotes, company.upper())
    
    years = [2014,2013, 2012, 2011, 2010]  #the year to print
    #plot the events along with the tweets volume and price
    for year in years:
        plot_by_event_year("%s tweet volume vs Quote" % company.upper(), tweet_days, tweet_counts, events, quotes, year, "quote")
            
    #plot the events along with the tweet volume and zscore30
    for year in years:
        plot_by_event_year("%s tweet volume vs Quote_Zscore30" % company.upper(), tweet_days, tweet_counts, events, quote_z30, year, "zscore")
        
    #plot the events along with the retweet volume and price
    for year in years:
        plot_by_event_year("%s retweet volume vs Quote" % company.upper(), retweet_days, retweet_counts, events, quotes, year, "quote")
    
#     word_html = plot_words(tweets_files, 100)
    
    word_html = handle_word(events, tweets_files)
    return word_html

def extract_frequency(tweet_files, top_num):
    """Extracting the words frequency from tweets"""
    tokens = []
    for f in tweet_files:
        for l in open(f):
            t = json.loads(l)
            text = unicodedata.normalize('NFKC', t['text']).lower()
            tokens.extend(text.split(' '))
    tokens = [t.replace(",","").replace(".","").replace("\t","") for t in tokens if t != '' and len(t) != 1]
    words = pd.Series(tokens)
    
    #remove the word with length == 1
    ignore = set(stopwords.words('english')) & set(words.unique())
    
    relevant_words = words.value_counts().drop(ignore)
    single_word = set([k for k in relevant_words.keys() if len(k) == 1])
    relevant_words = relevant_words.drop(single_word)
    
    result = [(k,v) for k, v in relevant_words.head(top_num).iteritems() if len(k) != 0]
    return result

def plot_hisgram(words_count, in_title):
    fig = figure(figsize(20, 6), dpi=80)
    idx = range(len(words_count))
    entity_count = [v for k,v in words_count]
    ep = bar(idx, entity_count, color='y')
    words = [k for k,v in words_count]
    xticks(idx, words, rotation=70)
    xlabel("Words")
    ylabel("Frequency")
    title(in_title)

class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
        the form [[1,2,3],[4,5,6]], and renders an HTML Table in 
        IPython Notebook. """
    
    def _repr_html_(self):
        html = ["<table>"]
        for row in self:
            html.append("<tr>")
            
            for col in row:
                html.append("<td>{0}</td>".format(col))
            
            html.append("</tr>")
        html.append("</table>")
        return ''.join(html)

def plot_words(tweets_files, top_num):
    rewords = extract_frequency(tweets_files, top_num)  
    plot_hisgram(rewords, 'Top 100 Words Frequency in the Tweets related SAPE From 2009-01-01 to 2014-05-02')
    table = ListTable()
    table.append(["Words", "Frequency"])
    for item in rewords:
        table.append(item)
    html = table._repr_html_()
    return html

def extract_tweets_around_impdays(event_days, tweet_files, window):
    def get_around_days(event_days, window):
        """compute the around days"""
        result = {}
        for day in event_days:
            around_days = [(datetime.datetime.strptime(day,"%Y-%m-%d") - datetime.timedelta(i)).strftime('%Y-%m-%d')
                           for i in range(1, window + 1)]
            result[day] = around_days
        return result
    a_days = get_around_days(event_days, window) #get the arounds
    
    dd_dmatch = {}
    for day, ds in a_days.items():
        for d_ds in ds:
            if d_ds not in dd_dmatch:
                dd_dmatch[d_ds] = []
            dd_dmatch[d_ds].append(day)           
    
    day_tweets = {}
    for f in tweet_files:
        with open(f) as tf:
            for l in tf:
                tweet = json.loads(l)
                dt = parser.parse(tweet['created_at']).strftime(date_format)
                if dt in dd_dmatch:
                    target_days = dd_dmatch[dt]
                    for target_day in target_days:
                        if target_day not in day_tweets:
                            day_tweets[target_day] = {}
                        if dt not in day_tweets[target_day]:
                            day_tweets[target_day][dt] = []
                        day_tweets[target_day][dt].append(tweet)
                    
    return day_tweets
                
def count_frequency(tweets, k):
    """count the frequency of tweets"""
    final_tokens = []
    for t in tweets:
        text = unicodedata.normalize('NFKC', t['text']).lower() #get the content
        text = re.sub(r'[.,?!()]', '', text) # remove the punctuation
        tokens = re.split(r'[ \t]+', text) # split by space
        tokens = [t for t in tokens if t not in stopwords.words('english') and len(t) !=0 and len(t)!=1]
#         tokens = nltk.bigrams(tokens)
        final_tokens.extend(tokens)
    result_fre = Counter(final_tokens).most_common(k)
    return result_fre

def filter_tweets_by_frq(tweets, frq_k):
    """filter the keywords by frq_k"""
    final_tokens = []
    for t in tweets:
        text = unicodedata.normalize('NFKC', t['text']).lower() #get the content
        text = re.sub(r'[.,?!()]', '', text) # remove the punctuation
        tokens = re.split(r'[ \t]+', text) # split by space
        tokens = [t for t in tokens if t not in stopwords.words('english') and len(t) !=0 and len(t)!=1]
        final_tokens.extend(tokens)
    word_frq = Counter(final_tokens).items()
    result = [(word, frq) for word, frq in word_frq if frq >= frq_k]
    return result

def get_tweets_frq(day_tweets):
    """get the daily frequency"""
    tweets_frq = {}
    for eventDay, day_infos in day_tweets.items():
        tweets_frq[eventDay] = {}
        for day, tweets in day_infos.items():
            day_tokens = []
            for t in tweets:
                text = unicodedata.normalize('NFKC', t['text']).lower() #get the content
                text = re.sub(r'[.,?!()]', '', text) # remove the punctuation
                tokens = re.split(r'[ \t]+', text) # split by space
                tokens = [t for t in tokens if t not in stopwords.words('english') and len(t) !=0 and len(t)!=1]
                day_tokens.extend(tokens)
            word_frq = Counter(day_tokens).items()
            tweets_frq[eventDay][day] = word_frq
    return tweets_frq

def filter_day_frq(tweets_frq, frq_k, num_day):
    """filter the keywords by frq_k"""
    result = {}
    new_result = {}
    for eventDay, day_info in tweets_frq.items():
        new_result[eventDay] = {}
        days = sorted(day_info.keys(), reverse=True)
        around_days = [(datetime.datetime.strptime(eventDay,"%Y-%m-%d") - datetime.timedelta(i)).strftime('%Y-%m-%d')
                           for i in range(1, num_day + 1)]
        tmp_day_frq = {}
        filter_words = []
        for day in around_days:
            if day not in day_info:
                continue
            frq = day_info[day]
            for k,v in frq:
                if k not in tmp_day_frq:
                    tmp_day_frq[k] = 0
                tmp_day_frq[k] += v
        filter_words = [word for word, count in tmp_day_frq.items() if count >=frq_k]
        
        
        #compuate the standard deviation
        for word in filter_words:
            word_list = {}
            for day in days:
                frq = day_info[day]
                for k,v in frq:
                    if k == word:
                        word_list[day] = v
                        break
                
            new_result[eventDay][word] = word_list
        result[eventDay] = filter_words

    
    return result, new_result

def handle_word(events, tweet_files):
#     events = get_custormized_days("sape")
#     tweet_files = glob('/home/weiwang/workspace/data/topsy_50/SAPE*')
    day_tweets = extract_tweets_around_impdays(events,tweet_files,30)
    tweet_frq = get_tweets_frq(day_tweets)
    result, new_result = filter_day_frq(tweet_frq, 5, 3)
    rank = {}
    for event_day, word_info in new_result.items():
        #fullfill the missdays as zero
        around_days = [(datetime.datetime.strptime(event_day,"%Y-%m-%d") - datetime.timedelta(i)).strftime('%Y-%m-%d')
                               for i in range(1, 30 + 1)]
        word_zscore = []
        for word, word_day_info in word_info.items():
            count = [word_day_info.get(day,0) for day in around_days]
            dev = np.std(count[3:])
            if dev == 0:
                dev = 1
            zscore = max([(count[i] - np.mean(count[3:]))/dev for i in range(3)])
            word_zscore.append((word, zscore))
        word_zscore.sort(key=lambda x:x[1], reverse=True)
        
        
        tmp = ["%s   %.2f" % (k,v) for k,v in word_zscore[:30] if v > 2]
        if len(tmp) < 30:
            tmp.extend([""]*(30-len(tmp)))
        rank[event_day] = tmp
        
    return pd.DataFrame(rank).to_html()
    

"""
The day_tweets format:
    {
     "eventDay1": {
                   "d1": [t1, t2, t3],
                   "d2": [t4, t5, t6]
                   }
     }
The keyword frequency format:
    {
     "eventDay1":
    {
          "d1":{w1:c1, w2:c2},
          "d2":{w3:c3, w4:c4}
      }
     
     }
"""
```
