```python

"""
Extract import days from EDGAR:
consists of 10-Q(quarterly report), 10-K(Annual report), 8-K(important event)
"""
import urllib2
import urllib
from bs4 import BeautifulSoup

byd_url = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0000906553&type=8-K%25&dateb=&owner=exclude&count=40"
BASIC_URL = "http://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=%s&type=%s&dateb=&owner=exclude&count=40"
def get_days(soup):
    obs = soup.find_all(class_="tableFile2")[0].find_all('tr')
    days = [o.find_all('td')[3].text for o in obs[1:]]
    return sorted(days, reverse=True)

def ingest_event_days(url):
    event_days = []
    soup = BeautifulSoup(urllib2.urlopen(url))
    next_flag = soup.find_all("input", attrs={"value": "Next 40"})
    start = 0
    event_days = event_days + get_days(soup)
    while len(next_flag) > 0:
        start = start + 40
        next_url = "%s&start=%d" % (url, start)
        soup = BeautifulSoup(urllib2.urlopen(next_url))
        event_days = event_days + get_days(soup)
        next_flag = soup.find_all("input", attrs={"value": "Next 40"})
    return event_days

def ingest(cik):
    day_events = {}
    event_types = ["8-K", "10-Q", "10-K"]
    for event_type in event_types:
        url = BASIC_URL % (cik, event_type)
        days = ingest_event_days(url)
        day_events.update({d:event_type for d in days})
    return day_events
        
byd_events = ingest("0000906553")
```

```python
from dateutil import parser
import sys
from datetime import timedelta
from glob import glob
import json
"""
work steps:
1> getting the stock price of these companies from yahoo
2> get the annual report date of these companies from EDGAR
3> compute the daily tweet count for these companies
4> find the entities in the tweets
5> find who are the guies most tweet about these companies
6> which tweets of these companies has been retweeted many times(the retweets also capture some events, but it seems that 
people people only talk it and in the same day)
7> compute the zscore of the stock market to check the 
"""

mpwr_files = glob("/home/weiwang/workspace/data/topsy_50/MPWR*")
byd_files = glob("/home/weiwang/workspace/data/topsy_50/BYD*")
sape_files = glob("/home/weiwang/workspace/data/topsy_50/SAPE*")


def load_tweets_by_day(twteet_files):
    """
    Group the tweets day by day
    args:
        tweet_files:  the tweet files list
    returns:
        tweets:  the tweets dictionary, key is date and value is the tweets list
    """
    tweets = {}
    for f in twteet_files:
        with open(f) as tf:
            for l in tf:
                try:
                    t = json.loads(l)
                    dt = parser.parse(t['created_at']).strftime('%Y-%m-%d')
                    if dt in tweets:
                        tweets[dt].append(t)
                    else:
                        tweets[dt] = [t]
                except:
                    print "Error:[%s]\n%s" % (sys.exc_info()[0], l)
    return tweets

def get_tweets_by_range(tweets, start, end):
    """"
    Get the tweets in a date range
    args:
        tweets:  the original tweets dictionary
        start: the start date
        end:  the end date
    returns:
        the tweets filtered by date range
    """
    return {k:v for k,v in tweets.items() if k >= start and k <= end}

def get_tweets_by_windows(tweets, day, window):
    """"
    Get the tweets in a date window
    args:
        tweets:  the original tweets dictionary
        day: the indicator date
        window: the size of the time window, it can be positive or negative
    returns:
        the tweets filtered by date window
    """
    second_date = (parser.parse(day) - timedelta(days=window)).strftime("%Y-%m-%d")
    if day > second_date:
        start = second_date
        end = day
    else:
        start = day
        end = second_date
    return get_tweets_by_range(tweets, start, end)
    
#byd_tweets = load_tweets_by_day(byd_files)
```

```python
import nltk
import pandas as pd
import unicodedata
from collections import Counter
from pytagcloud import create_tag_image, make_tags
from pytagcloud.lang.counter import get_tag_counts
from IPython.core.display import Image 

def make_tag_cloud(freqs):
    tags = make_tags(freqs)
    fname = 'noun_last_words.png'
    create_tag_image(tags, fname, size=(700, 400), fontname='Droid Sans')
    return fname

def content_analyze(tweets):
    """
    count the words frequency without stop words
    """
    tokens = []
    for t in tweets:
        text = unicodedata.normalize('NFKC', t['text']).lower()
        tokens.extend([w for w in text.split() if w not in ['-', 'rt', '...', 'boyd', 'gaming', '$byd'] + nltk.corpus.stopwords.words('english')])
    token_counter = Counter(tokens)
    common = token_counter.most_common(100)
    imgpath = make_tag_cloud(common)
    display(Image(filename=imgpath)) 

def author_analyze(tweets, top):
    df = pd.DataFrame(tweets)
    authors = df.pop('user').apply(pd.Series)
    author_count = Counter(authors.screen_name)
    return author_count.most_common(top)

def entities_analyze(tweets, top):
    entities = []
    for t in tweets:
        t_entities = t.get('entities', {})
        entities.extend(["@%s" % u["screen_name"] for u in t_entities.get('user_mentions', [])])
        entities.extend([u["url"] for u in t_entities.get('urls', [])])
        entities.extend(["#%s" % u["text"] for u in t_entities.get('hashtags', [])])
        
    entity_counter = Counter(entities)
    top_k = entity_counter.most_common(top)
    if len(top_k) < top:
        top_k.extend([("",0)] * (top - len(top_k)))
    return top_k
        
def batch_analysis(quarterly_days, dic_tweets, analysis_func, top):
    rank = {}
    for day in quarterly_days:
        quarter_tweets = get_tweets_by_windows(dic_tweets, day, 15)
        tweets = []
        for d in quarter_tweets:
            tweets.extend(quarter_tweets[d])
        comm = analysis_func(tweets, top)
        rank[day] = ["%s (%s)" % (q[0], q[1]) for q in comm]
    return rank

#author_rank = batch_analysis(quarterly_days, byd_tweets, author_analyze, 15)       
entities_rank = batch_analysis(all_event_days, byd_tweets, entities_analyze, 10)       
#entities_analyze()    
```
